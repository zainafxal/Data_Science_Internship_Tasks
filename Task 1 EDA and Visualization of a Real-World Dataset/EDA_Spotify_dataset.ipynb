## **6.0 Data Cleaning**

### **6.1 Removeing "Unnamed: 0" & "track_id" Columns**

* We removed `Unnamed: 0` & `track_id` columns to focus on relevant data for understanding Popular Songs behavior in Spotify Charts.

```python
df.drop(columns=['Unnamed: 0', 'track_id'], inplace=True)
```

### **6.2 mode column**

```python
# Check missing values first in mode column
df["mode"].isnull().sum()
```

### Observation: 👀
As there is no missing values in `mode` column,So we are going to rename this column name to `scale` & data type will be converted from int to Object (from 0/1 to 'Major' for 1 and 'Minor' for 0) .

### 6.2.1 Rename and convert data type of mode column

```python
df['scale'] = df['mode'].replace({0: 'Minor', 1: 'Major'}).astype('object')
df.drop(columns=['mode'], inplace=True) # Drop the original 'mode' column
```

* Now we have renamed the mode column as scale and also converted the data type of column from int to object (0/1 to 'Minor' and 'Major')

### **6.3 key column**

```python
# Check missing values in key column
df["key"].isnull().sum()
```

### Observation: 👀
As there is no missing values in `key` column,So we are going to convert data type from int to object and values (0,1,-1) of column into (G#,B,A etc) for better Understanding.

### 6.3.1 Conversion of data type and Values of key column

```python
# List of musical keys
keys = [
    'C', 'C♯/D♭', 'D', 'D♯/E♭', 'E', 
    'F', 'F♯/G♭', 'G', 'G♯/A♭', 'A', 
    'A♯/B♭', 'B', 'No key detected'
]

# Replace integers with corresponding musical keys
df['key'] = df['key'].replace(range(-1, 12), keys)
```

* Now we have converted all the int values into strings with respective key values hence data type of the column is also changed.

## **11.0 Outliers Detection**

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Identify outliers using the IQR method
outliers_count = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_count[col] = len(outliers)


# boxplot for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

```

### Outliers count per numerical column:

```python
# Print the number of outliers per column
for col, count in outliers_count.items():
    print(f"{col}: {count}")
```


# Understanding Outliers

**Duration_ms:**  
 * Very long songs could be extended versions, experimental tracks, or even mislabeled podcast episodes.

**Danceability, Energy, Loudness, Valence:**

**Speechiness:**  
 * High speechiness indicates spoken word content (podcasts, audiobooks). These are clearly different from music.

**Acousticness, Instrumentalness:**  
 * Outliers here might be very electronic or very acoustic tracks, or purely instrumental pieces.

**Liveness:**  
 * High liveness suggests live recordings, which naturally differ from studio tracks.

**Tempo:**

## **12.0 📊 Visualizations**

### 12.1 Categorical variables

```python
# Set style
sns.set_style('whitegrid')

# Categorical variables (top 10 frequent values)
categorical_cols = ['track_genre', 'artists', 'album_name', 'track_name']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    top_10 = df[col].value_counts().head(10)
    top_10_df = top_10.reset_index()
    top_10_df.columns = [col, 'count']
    sns.barplot(x=col, y='count', data=top_10_df, palette='Set2')
    plt.title(f'Top 10 {col}', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

### 12.2 Numeric variables

```python
# Set style
sns.set_style('whitegrid')

# Numeric Distributions (each in separate plots)
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30, color='teal')
    plt.title(f'Distribution of {col}', fontsize=16)
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

### 12.3 Correlation Heatmap

```python
# Correlation Heatmap
plt.figure(figsize=(16, 12))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

## 12.4 **Histograms**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dataset.csv')

# Set up the histogram parameters
features = ['danceability', 'energy', 'acousticness', 'valence', 'tempo', 'duration_ms']
plt.figure(figsize=(15, 10))

# Create a histogram for each feature
for i, feature in enumerate(features):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[feature], bins=30, color='blue', alpha=0.7)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()
```
